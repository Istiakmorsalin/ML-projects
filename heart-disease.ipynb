```python
import pandas as pd
import numpy as np
heart_disease = pd.read_csv("heart-disease.csv") 
heart_disease
```

```python
X = heart_disease.drop("target", axis = 1)  
y = heart_disease["target"]
```

```python
X
```

```python
from sklearn.ensemble import RandomForestClassifier  
clf = RandomForestClassifier(n_estimators=100) 

clf.get_params()
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2) 
X_train
```

```python
clf.fit(X_train, y_train)
```

```python
y_preds = clf.predict(X_test) 
y_preds
```

```python
y_test
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score  
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds) 
```

```python
np.random.seed(42)    
for i in range(10,100,10):  
    print("Trying model with {i} estimators...")  
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train) 
    print(f"Model accurary on test set: {clf.score(X_test, y_test) * 100}%")
    print("")
```

```python
import pickle  
pickle.dump(clf, open("random_forest_model1_1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model1_1.pkl", "rb"))
```

```python
loaded_model.score(X_test, y_test)
```
